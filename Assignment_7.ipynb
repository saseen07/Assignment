```python
#1. Create an array of shape (100,100,3) of unsigned integer of 8 bits (uint8) filled with random values between 0 to 256.
#2. Find the values that are less than 30, 50, 100, 250.
#3. Copy and change the value that are less than 110 to 0 and others to 255.
#4. Split the array into 4 parts vertivally and horizontally
#5. Hint use random seed of 12345
```

```python
def array_properties(a):
    print("a = \n", a)
    print("dim = ", a.ndim)
    print("shape = ", a.shape)
    print("datatype = ", a.dtype)
    print("size = ", a.size)
    print()
```

```python
import numpy as np
np.random.seed(12345)
arr = np.random.randint(0,256,(100,100,3),dtype=np.uint8)
array_properties(arr)
```

```python
bool30 = arr <30
array_properties(bool30)
```

```python
bool30 = bool30 == True
array_properties(bool30)

result_30 = arr[bool30]
print("Values less than 30 are :")
array_properties(result_30)
```

```python
bool50 = arr <50
array_properties(bool50)
```

```python
bool50 = bool50 == True
array_properties(bool50)

result_50 = arr[bool50]
print("Values that are less than 50 are :")
array_properties(result_50)
```

```python
bool100 = arr <100
array_properties(bool100)
```

```python
bool100 = bool100 == True
array_properties(bool100)

result_100 = arr[bool100]
print("Values that are less than 100 are :")
array_properties(result_100)
```

```python
bool250 = arr <250
array_properties(bool250)
```

```python
bool250 = bool250 == True
array_properties(bool250)

result_250 = arr[bool250]
print("Values that are less than 250 are :")
array_properties(result_250)
```

```python
arr_c = np.copy(arr)
```

```python
bool110 = arr_c <110
array_properties(bool110)
```

```python
bool0 = bool110 == True
bool255 = bool110 == False
array_properties(bool110)


arr_c[bool0] = 0
arr_c[bool255] = 255

array_properties(arr_c)
```
